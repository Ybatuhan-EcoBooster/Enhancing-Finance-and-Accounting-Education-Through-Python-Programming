```python
def Yahoo_Finance(ticker,period = None,start_date = None, end_date = None):

        Company_ticker = yf.Ticker(ticker)

        if period is not None:
            Data = Company_ticker.history(period= period)
            

        elif start_date is not None:
            Data = yf.download(Company_ticker.ticker, start= start_date)

        elif end_date is not None:
            Data = yf.download(Company_ticker.ticker, end = end_date)

        elif start_date and end_date is not None:
            Data = yf.download(Company_ticker.ticker, start= start_date, end= end_date) 
        
        else:
            Data = yf.download(Company_ticker.ticker, period="max")

        def majorholder():
            majorholder = Company_ticker.major_holders
            return majorholder

        major = majorholder()

        return Data, major

    
```

## Yahoo Finance Important Functions

```python
# MCD = McDonald's Corporation 

MCD = yf.Ticker("MCD")

info_keys = MCD.info.keys()          # Company information

MCD_info = []

for i in info_keys:
    v = MCD.info.get(i)
    MCD_info.append(v)
MCD_info = pd.DataFrame(MCD_info, index = info_keys, columns=["Information"])
MCD_info.index.name = "Title"

MCD_info
```

```python
def news(self):
        self = yf.Ticker(self)
        news = self.news
        news = pd.DataFrame(news)
        news = news[["title","publisher","link"]]
        news.columns = ["Title","Publisher","News Link"]

        return news
```
